```python
import numpy as np

eig = np.linalg.eig
matrix = np.matrix
testmatrix = matrix([[4,2],[2,4]]) #메트릭스 생성

print(eig(testmatrix)) # Eigenvalue와 Eigenvector
```

```python
value = eig(testmatrix)[0] #Eigenvalue
vector = eig(testmatrix)[1] #Eigenvector

print('Eigenvalue :', value)
print('Eigenvalue :', vector)
```

```python
testmatrix = matrix([[4,2],[3,5]])
value = eig(testmatrix)[0] #Eigenvalue
vector = eig(testmatrix)[1] #Eigenvector

print('Eigenvalue :', value)
print('Eigenvalue :', vector)
```

```python
import numpy as np
A = np.array([[-2,-5,-2],[-5,1,-5],[-2,-5,2]])
print("Matrix A: ")
print(A)
print()

eigenvalue, eigenvetor = np.linalg.eig(A)
eigenvalue = np.round(eigenvalue, 5) #소수점 5자리로 반올림
eigenvetor = np.round(eigenvetor, 5)
print("Eigen Value: ")
print(eigenvalue)
print()

print("Eigen Value Diagonal Matrix \u039B(lambda):")
eigenvalue = np.array([[eigenvalue[0],0,0],[0, eigenvalue[1], 0], [0,0,eigenvalue[2]]])
print(eigenvalue)
print()

print("Eigen Vetor Matrix P: ")
print(eigenvetor)
print()

print("AP = ")
print(np.round(A@eigenvetor,0))
print()

print("P\u039B =")
print(np.round(eigenvetor@eigenvalue))
```
