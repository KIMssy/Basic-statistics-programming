```python
#결측치 처리 함수(평균사용)
def missing_data(y1,y2):
  avg_data = (y2+y1)/2
  return avg_data
```

```python
y[1]= missing_data(y[0], y[2])
plt.grid(color="green", alpha=0.5, linestyle= '--')
plt.plot(x,y, marker = 'o')
```
